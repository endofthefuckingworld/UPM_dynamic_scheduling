```python
import numpy as np
import pandas as pd
```

```python
n_jobs = 15
n_machines = 4
k = 2.5

setup_times = np.random.randint(1,20,[n_jobs, n_jobs])

job_data = np.zeros((n_jobs,4))
for i, job in enumerate(job_data):
    job[0] = i+1  #type
    job[1] = np.random.randint(1, 10*n_jobs/n_machines) #arrival_time
    job[2] = np.random.randint(10, 100)  #process_time
    job[3] = job[1] + k* job[2]
    

job_data_df = pd.DataFrame(job_data, columns = ['type','arrival time','process time','due dates']).sort_values(by=['arrival time'])
setup_times_df = pd.DataFrame(setup_times)

display(job_data_df)
display(setup_times_df)
job_data_df.to_csv('job_data.csv', index=False)
setup_times_df.to_csv('setup_times.csv',  index=False)
```

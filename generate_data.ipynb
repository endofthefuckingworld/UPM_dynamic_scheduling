```python
import numpy as np
import pandas as pd
```

```python
n_jobs = 15
n_machines = 4
k = 2.5

for n in range(30):
    setup_times = np.random.randint(1,20,[n_jobs, n_jobs])
    for i in range(n_jobs):
        setup_times[i,i] = 0

    Columns = ['type','arrival time']
    job_data = np.zeros((n_jobs,n_machines+3))

    for i, job in enumerate(job_data):
        job[0] = i+1  #type
        job[1] = np.random.randint(1, 10*n_jobs/n_machines) #arrival_time
        process_time = np.random.randint(10, 100)
        for j in range(n_machines):
            job[2+j] = np.random.normal(process_time, 2)  #process_time
        job[2+n_machines] = job[1] + k* process_time

    for i in range(n_machines):
        Columns.append('process_time'+str(i))
    Columns.append('due dates')

    job_data_df = pd.DataFrame(job_data, columns = Columns).sort_values(by=['arrival time'])
    setup_times_df = pd.DataFrame(setup_times)

    display(job_data_df)
    display(setup_times_df)
    job_data_df.to_csv('Experiment_data/job_data'+str(n)+'.csv', index=False)
    setup_times_df.to_csv('Experiment_data/setup_times'+str(n)+'.csv',  index=False)
```
